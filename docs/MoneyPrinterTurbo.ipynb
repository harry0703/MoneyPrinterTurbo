## 2. Configurar o ngrok para Acesso Remoto

Usaremos o ngrok para criar um túnel seguro e expor o nosso servidor Streamlit local na internet.

**Importante**: Precisa de obter o seu token de autenticação no [painel do ngrok](https://dashboard.ngrok.com/get-started/your-authtoken) para usar este serviço.

```python
from pyngrok import ngrok

# Terminar qualquer túnel ngrok existente
ngrok.kill()

# Definir o seu token de autenticação
# Substitua "your_ngrok_auth_token" pelo seu token real
ngrok.set_auth_token("your_ngrok_auth_token")
```

## 3. Iniciar a Aplicação e Gerar URL Pública

Agora vamos iniciar o servidor Streamlit e criar o túnel ngrok para torná-lo acessível online:

```python
import subprocess
import time

print("🚀 A iniciar o MoneyPrinterTurbo...")
# Entramos no .venv do projeto através de uv run, garantindo que o Streamlit utilize as versões de dependências bloqueadas no passo anterior.
streamlit_proc = subprocess.Popen([
    "uv", "run", "streamlit", "run", "./webui/Main.py", "--server.port=8501", "--server.address=0.0.0.0"
])

# Aguardar inicialização do servidor
time.sleep(5)

print("🌐 A criar túnel ngrok para expor o MoneyPrinterTurbo...")
public_url = ngrok.connect(8501, bind_tls=True)

print("✅ Implantação concluída! Aceda ao seu MoneyPrinterTurbo em:")
print(public_url)
```
